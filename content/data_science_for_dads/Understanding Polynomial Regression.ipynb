One of the things that always drove me a bit crazy was when I heard people talk about linear regression working for finding polynomial (i.e., non-linear) coefficients. 

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
```

First step is to create some data with a bit of noise 

```python
x = np.linspace(1, 10, 100)
y = np.random.normal(2 * x**4, 1000)
```

```python
m = np.polyfit(x**4, y, deg=1)[0]
print('Slope (m): {0:.2f}'.format(m))
```

Notice that we are still doing a normal linear regression. In fact, we have set the degree of the polynomial that we are solving for to $1$, so we really are trying to find a line that best fits the data. In order to do that, we have to first tranform the data by raising our input to the power of $4$. Our regression finds a slope of $1.97$, which is quite close to the value of the actual coefficient. 

```python
plt.gcf().set_figwidth(12)

# Plot the data as is so we can see its natural curvature
plt.subplot(131)
plt.scatter(x, y, alpha=0.7)

# Plot the data 
plt.subplot(132)
plt.scatter(x**4, y, alpha=0.7)
plt.plot(x**4, m*x**4, color='red', linewidth=2)
plt.tick_params(labelleft=False, left=False)

plt.subplot(133)
plt.scatter(x, y, alpha=0.7)
plt.plot(x, m*x**4, color='red', linewidth=2)
plt.tick_params(labelleft=False, left=False)

plt.tight_layout()
```
